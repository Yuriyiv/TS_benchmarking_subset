```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# cur_dir = os.getcwd().split('/')
# if cur_dir[-1] == 'notebooks':
#     os.chdir("..")
```

#### Simple features

```python
df_features = pd.read_csv("features.csv", index_col=0)
df_features.columns
```

```python
df_features.describe().iloc[1:,:].apply(lambda x: np.round(x, 2))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes = axes.flatten()

sns.kdeplot(df_features['entropy'], ax=axes[0], fill=True, color='black')
axes[0].set_title('KDE Plot of Entropy')

sns.kdeplot(df_features['gini'], ax=axes[1], fill=True, color='brown')
axes[1].set_title('KDE Plot of Gini')

sns.kdeplot(df_features['number_of_classes'], ax=axes[2], fill=True, color='red')
axes[2].set_title('KDE Plot of Number of Classes')

sns.kdeplot(df_features['size'], ax=axes[3], fill=True, color='chocolate')
axes[3].set_title('KDE Plot of Size')

plt.tight_layout()

plt.show()
```
